# Word Anagram Challenge
- [x] Read in a list of words from a file
- [x] separate words into word classes by same length
- [ ] For each class of words of same length find anagrams
- [ ] Count the total number of anagrams in each class

```python
#read words from file into list
uniqueWords = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
import collections
sameLength = collections.defaultdict(list)
for word in uniqueWords:
    sameLength[len(word)].append(word)
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
wordBySignature = collections.defaultdict(list)
for word in uniqueWords:
    wordBySignature[signature(word)].append(word)
def anagram_fast(word):
    return wordBySignature[signature(word)]
anagram_fast('love')
```

```python
#find anagrams in the same length
anagrams_sameLength = {}
for length, words in sameLength.items():
    anagrams_sameLength[length] = {word: anagram_fast(word) for word in words if len(anagram_fast(word)) >1}
anagrams_sameLength[3]
```

```python
#find count of anagrams in the same length
anagrams_sameLength = {}
for length, words in sameLength.items():
    #only want the # of anagrams -1 since it counts itself
    anagrams_sameLength[length] = {word: len(anagram_fast(word)) - 1 for word in words if len(anagram_fast(word)) >1}
anagrams_sameLength[3]
```

```python
#find sum of each class
anagrams_sameLength = {}
for length, words in sameLength.items():
    #only want the # of anagrams -1 since it counts itself
    anagrams_sameLength[length] = sum(len(anagram_fast(word)) - 1 for word in words if len(anagram_fast(word)) >1)/2
anagrams_sameLength
```

```python
max(anagrams_sameLength)
```
